```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("file\\HousePrice.csv")
df
```

```python
x = df.iloc[:,0].values.reshape(-1,1)
x
```

```python
y = df.iloc[:,1].values
y
```

```python
plt.scatter(x, y)
plt.title('House Price by Size.')
plt.xlabel('House Area(in square meter)')
plt.ylabel('Selling Price(in million PHP)')
```

```python
lr = LinearRegression()
model = lr.fit(x, y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state = 0)
x,y
```

```python
print('x training data:\n', x_train)
print('y training data:\n', y_train)
```

```python
plt.scatter(x_train, y_train)
plt.xlabel('Area in SQM')
plt.ylabel('Price in milion PHP')
```

```python
plt.scatter(x_test, y_test)
plt.xlabel('Area in SQM')
plt.ylabel('Price in milion PHP')
```

```python
model = lr.fit(x_train, y_train)
model.score(x_train, y_train)
```

```python
y_train_pred = model.predict(x_train)
```

```python
y_train_pred, y_train, y
```

```python
plt.plot(x_train, y_train, 'o', color='green')
plt.plot(x_train, y_train_pred, color='orange')
plt.title('House Price by Size.')
plt.xlabel('House Area(in square meter)')
plt.ylabel('Selling Price(in million PHP)')
```

```python
y_test_pred = model.predict(x_test)
y_test_pred, y_test
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.metrics import mean_absolute_error
mae_train = mean_absolute_error(y_train, y_train_pred)
print('training data MAE: ', mae_train)
```

```python
mae_test = mean_absolute_error(y_test, y_test_pred)
print('training data MAE: ', mae_test)
```
