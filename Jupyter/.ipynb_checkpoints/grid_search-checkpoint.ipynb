```python
from sklearn import svm, datasets
iris = datasets.load_iris()
dir(iris)
iris.target_names
```

```python
import pandas as pd
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['flower'] = iris.target
df['flower'] = df['flower'].apply(lambda x: iris.target_names[x])
df[47:52]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2)
```

```python
model = svm.SVC(kernel='rbf', C=30, gamma='auto')
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(svm.SVC(kernel='linear', C=10, gamma='auto'), iris.data, iris.target, cv=5)
```

```python
cross_val_score(svm.SVC(kernel='rbf', C=10, gamma='auto'), iris.data, iris.target, cv=5)
```

```python
cross_val_score(svm.SVC(kernel='linear', C=20, gamma='auto'), iris.data, iris.target, cv=5)
```

```python
import numpy as np
kernels = ['rbf', 'linear']
C = [1, 10, 20]
avg_score = {}
for kval in kernels:
    for cval in C:
        cv_scores = cross_val_score(svm.SVC(kernel=kval, C=cval, gamma='auto'), iris.data, iris.target, cv=5)
        avg_score[kval + '_' + str(cval)] = np.average(cv_scores)
avg_score
```

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(svm.SVC(gamma='auto'),{
    'C': [1, 10, 20],
    'kernel': ['rbf', 'linear']
}, cv=5, return_train_score=False)
clf.fit(iris.data, iris.target)
clf.cv_results_
```

```python
df_result = pd.DataFrame(clf.cv_results_)
df_result
```

```python
df_result[['param_C', 'param_kernel', 'mean_test_score']]
```

```python
dir(clf)
```

```python
clf.best_score_
```

```python
clf.best_params_
```

```python
from sklearn.model_selection import RandomizedSearchCV
rs = RandomizedSearchCV(svm.SVC(gamma='auto'),{
        'C': [1, 10, 20],
        'kernel': ['rbf', 'linear']
    },
    cv= 5,
    return_train_score=False,
    n_iter=2
)
rs.fit(iris.data, iris.target)
pd.DataFrame(rs.cv_results_)[['param_C', 'param_kernel', 'mean_test_score']]
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
model_params = {
    'svm': {
        'model': svm.SVC(gamma='auto'),
        'params': {
            'C': [1, 10, 20],
            'kernel': ['rbf', 'linear']
        }
    },
    'Random_forest': {
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators': [1, 5, 10]
        }
    },
    'Logistic_Regression': {
        'model': LogisticRegression(solver='liblinear', multi_class='auto'),
        'params': {
            'C': [1, 5, 10]
        }
    }
}
```

```python
a = GridSearchCV(svm.SVC(gamma='auto'), {'C': [1, 10, 20], 'kernel': ['rbf', 'linear']}, cv=5, return_train_score=False)
a.fit(iris.data, iris.target)
```

```python
b = GridSearchCV(RandomForestClassifier(), {'n_estimators': [1, 5, 10]}, cv=5, return_train_score=False)
b.fit(iris.data, iris.target)
```

```python
c = GridSearchCV(LogisticRegression(solver='liblinear'),{'C': [1, 5, 10]}, cv=5, return_train_score=False)
c.fit(iris.data, iris.target)
```

```python
score = []

for model_name, mp in model_params.items():
#     print(mp['model'])
#     print(mp['params'])
    clf = GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
    clf.fit(iris.data, iris.target)
    score.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
pd.DataFrame(score)
```
