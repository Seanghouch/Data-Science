```python
import requests
import pandas as pd

url = "https://api.nationalize.io/?name=nathaniel"
response = requests.get(url)
if response.status_code == 200:
      data = response.json()
else:
  print("Error:", response.status_code)
df = pd.DataFrame(data['country'])
df.head()
```

```python
%matplotlib inline
df.plot(y="probability", x="country_id", color="Green")
```

```python
import pandas as pd

df = pd.DataFrame({
    "Date": pd.to_datetime(["2022-01-01", "2022-01-02", "2022-01-03", "2022-01-04", "2022-01-05"]),
    "Open": [100, 101, 102, 103, 104],
    "High": [90, 76, 89, 65, 112],
    "Low": [90, 96, 73, 54, 87],
    "Close": [103, 104, 105, 106, 107]
})
print(df)
df.plot(y=["Low","Open", "High", "Close"], x="Date", color=["Orange", "Blue", "Green", "Red"])
```
