```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
plt.legend(loc='upper right')
```

```python
# set axes labels and chart title
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')


plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
plt.legend(loc='best')
plt.grid()
```

```python
company=['Google', 'Amazon', 'Microsoft', 'Facebook']
revenue=[90, 136,89,27]
profit=[40, 2, 34, 12]
```

```python
xpos= np.arange(len(company))
xpos
```

```python
plt.xticks(ypos, company)
plt.ylabel('Revenue(bln)')
plt.title('US Tech Stocks')
plt.bar(xpos-0.2, revenue, width=0.4, label='Revenue')
plt.bar(xpos+0.2, profit, width=0.4, label='Profit')
plt.legend()
```

```python
plt.yticks(ypos, company)
plt.title('US Tech Stocks')
plt.barh(xpos-0.2, revenue, height=0.4, label='Revenue')
plt.barh(xpos+0.2, profit, height=0.4, label='Profit')
plt.legend()
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.xlabel('sugar range')
plt.ylabel('Total no of patients')
plt.title('Total no of patients')
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green', 'orange'],
         label=['men', 'women'], orientation='horizontal')
plt.legend()
```
