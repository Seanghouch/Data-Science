```python
import pandas as pd

india_weather = pd.DataFrame({
    'city': ['mumbi', 'delhi', 'banglore'],
    'temparature': [32, 45, 30],
    'humidty': [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    'city': ['new york', 'chicago', 'orlando', 'a'],
    'temparature': [21, 14, 35, 12],
    'humidty': [68, 65, 75, 34],
    'a': [1,2,3,4]
})
us_weather
```

```python
df = pd.concat([india_weather, us_weather], ignore_index=True)
df.head(10)
```

```python
df = pd.concat([india_weather, us_weather], keys=['india', 'us'])
df
```

```python
df.loc['us']
```

```python
temparature_df = pd.DataFrame({
    'city': ['mumbi', 'delhi', 'banglore'],
    'temparature': [32, 45, 30]
})
temparature_df
```

```python
windspeed_df = pd.DataFrame({
    'city': ['delhi', 'mumbi', 'banglore', 'ss'],
    'windspeed': [7, 12, 9, 8.5]
})
windspeed_df
```

```python
new_df = pd.concat([temparature_df, windspeed_df], axis=1)
new_df
```

```python
df1 = pd.DataFrame({"key": ["a", "b", "c","d"], "value": [1, 2, 3, 10]})
df2 = pd.DataFrame({"key": ["a", "b", "d"], "value": [4, 5, 7]})

df_merged = df1.merge(df2, on="key")

df_merged
```

```python
df4 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'], 'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})
other = pd.DataFrame({'key': ['K1', 'K0', 'K2'], 'B': ['B0', 'B1', 'B2']})
df5 = df4.set_index('key').join(other.set_index('key'))
df5
```
