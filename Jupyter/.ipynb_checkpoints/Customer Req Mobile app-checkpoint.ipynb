```python
import decimal
import pandas as pd

df = pd.read_excel('Mobile App Customer.xlsx')
data = []
print(df)

```

```python
df.shape[0]
```

```python
for i in range(df.shape[0]):
    no = df['NO'][i]
    conducted_branch = df['CONDUCTED BRANCH'][i]
    cif_number = df['CIF NUMBER'][i]
    customer_name = df['CUSTOMER NAME'][i]
    date_time = df['DATE AND TIME'][i]
    precessed_by = df['PROCESSED BY'][i]
    approved_by = df['APPROVED BY'][i]
    sub_df1 = pd.DataFrame(df['EXISTING LIMIT'][i].split('\n'))
    sub_df2 = pd.DataFrame(df['NEW LIMIT'][i].split('\n'))
    for ii in range(sub_df1.shape[0]):
        if(len(sub_df1[0][ii]) != 0 or len(sub_df2[0][ii]) != 0):
            split_string_existing = sub_df1[0][ii].split('-')
            transaction_name_existing = split_string_existing[0].strip()
            transaction_value_existing = split_string_existing[1].strip()
            # print(transaction_value_existing)
            split_string_new = sub_df2[0][ii].split('-')
            transaction_name_new = split_string_new[0].strip()
            transaction_value_new = split_string_new[1].strip()
            transaction_value_new_n = float(transaction_value_new.replace(',', ''))
            transaction_value_existing_n = float(transaction_value_existing.replace(',', ''))
            status = ''
            difference_amount = float(transaction_value_new_n) - float(transaction_value_existing_n)
            if(difference_amount > 0):
                status = 'Increase'
            elif(difference_amount == 0):
                status = 'No Change'
            else:
                status = 'Decrease'
            data = data + [[no, conducted_branch, cif_number, customer_name, transaction_name_existing,
                            transaction_value_existing, transaction_name_new, transaction_value_new,
                            date_time, status, difference_amount, precessed_by, approved_by]]
new_data = pd.DataFrame(data, columns=['NO', 'CONDUCTED BRANCH', 'CIF NUMBER', 'CUSTOMER NAME',
                                       'EXISTING TRANSACTION NAME', 'EXISTING TRANSACTION VALUE',
                                       'NEW TRANSACTION NAME', 'NEW TRANSACTION VALUE', 'DATE AND TIME',
                                       'STATUS', 'DIFFERENCE AMOUNT', 'PROCESSED BY', 'APPROVED BY'])
new_data.head()
#pd.DataFrame(new_data).to_excel('Mobile App Customer v1.xlsx')
```

```python
new_data.info()
```

```python
to_drop = ['NO',
           'CONDUCTED BRANCH',
           'CIF NUMBER',
           'CUSTOMER NAME',
           'EXISTING TRANSACTION NAME',
           'EXISTING TRANSACTION VALUE',
           'NEW TRANSACTION VALUE',
           'DATE AND TIME',
           'PROCESSED BY',
           'APPROVED BY']

new_data.drop(to_drop, inplace = True, axis = 1)
new_data.head()
```

```python
pivot_table = new_data.pivot_table(values='DIFFERENCE AMOUNT', index='NEW TRANSACTION NAME', columns='STATUS', aggfunc='count')
df = pivot_table.fillna(0)

print(df.to_string(index=False))
```
