```python
import pandas as pd
import numpy as np
df = pd.read_csv('file\\carprices.csv')
df
```

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
merged = pd.concat([df, dummies], axis='columns')
merged
```

```python
final = merged.drop(['Car Model', 'Mercedez Benz C class'], axis='columns')
final
```

```python
x = final.drop(['Sell Price($)'], axis='columns')
x
```

```python
y = final['Sell Price($)']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
x_train
```

```python
x_test = np.array(x_test)
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
```

```python
clf.fit(x_test, y_test)
```

```python
clf.score(x_test, y_test)
```

```python
test_pred = clf.predict(x_test)
test_pred
```

```python
x_test_df = pd.DataFrame(x_test)
test_pred_df = pd.DataFrame(test_pred)
x_test_df
```

```python
test_pred_df
```

```python
new_df = pd.merge(x_test_df, test_pred_df, left_index=True, right_index=True)
new_df.columns = ['Mileage', 'Age(yrs)', 'Audi A5', 'BMW X5', 'Selling Price($)']
new_df
```

```python
merged = pd.concat([x_test_df, test_pred_df], axis='columns')
merged.columns = ['Mileage', 'Age(yrs)', 'Audi A5', 'BMW X5', 'Selling Price($)']
merged
```
